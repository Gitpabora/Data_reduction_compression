### K-mean steps :

 Step 1. An optimal number of cluster (K) is choosen.
 
 Step 2. k  number of points  "centroids" are initialized randomly within the data area.
 
 Step 3. Each data or observation is attributed to own closest centroid.
 
 Step 4. Updation is done for the centroids to hold the value corresponding to the center of  its all  attributed observations.
 
 Step 5. Steps 3-4  is repeated a number of times / until all of the centroids are prominent.

```python
import numpy as np
import matplotlib.pyplot as plt
img=plt.imread("/content/sample_image.jpg")

#Read and plot the image
plt.axis("off")
plt.imshow(img)
plt.show()

img_reshaped = np.reshape(img,(570,2955 ))

```

```python
img.shape
```

```python
Org_row = img. shape[0]
Org_col = img.shape[1]
```

```python

from sklearn.cluster import KMeans
```

```python
#Apply Kmean on the flattened image
image = img_reshaped

#Implement k-means clustering to form k clusters
kmeans = KMeans(n_clusters=64)
kmeans.fit(image)
```

```python
org_image_elt= image.shape[0]*image.shape[1]
max_cluster =100  ##experiment will be to check in intervals of 10
cp_ratio= [] 
cp_ratio = [0 for i in range(10)] ## intializing the compression ratio
```

reference :https://www.analyticsvidhya.com/blog/2021/01/in-depth-intuition-of-k-means-clustering-algorithm-in-machine-learning/

```python
for i in range( 10,max_cluster+10,10):
  kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
  pred_y = kmeans.fit_predict(image)
  plt.show()
  plt.title(" sctterplot for n_cluster is =%d" %i )
  plt.scatter(image[:,0], image[:,1])
  plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='red')
  print("when n_cluser =%d : " ,i) 
  print( kmeans.cluster_centers_.shape )
  new_image_elt = kmeans.cluster_centers_.shape[0]*kmeans.cluster_centers_.shape[1]
  cp_index = int(i/10) -1
  cp_ratio[cp_index]= ((org_image_elt-new_image_elt)/org_image_elt)*100
  print( "compression ratio (%)for n_cluser =%d is %d",i, cp_ratio[cp_index])
  fname_c =  "scatter_c" + str(i)
  plt.savefig(fname=fname_c, dpi=100)
  plt.show()
```
